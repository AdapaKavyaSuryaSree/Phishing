```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score as acs
```

```python
pData=pd.read_csv("/content/Phishing_Legitimate_full.csv")
print(pData.shape)
```

```python
pData.head()
```

```python
pData.isnull().sum()
```

```python
pData.CLASS_LABEL.value_counts()
```

```python
x=pData.drop(["id","CLASS_LABEL"],axis=1)
y=pData["CLASS_LABEL"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
xTrain,xTest,yTrain,yTest=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
xTrain.shape, yTrain.shape
```

```python
xTest.shape,yTest.shape
```

random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
Model=RandomForestClassifier(n_estimators=5000)
Model.fit(xTrain,yTrain)
```

```python
XTr=Model.predict(xTrain)
TrainingAcc=acs(yTrain,XTr)
TrainingAcc
```

```python
XTe=Model.predict(xTest)
TestingAcc=acs(yTest,XTe)
TestingAcc
```

xgboost

```python
from xgboost import XGBRegressor as XGB
Model1=XGB(n_estimators=1000,num_parallel_tree=10)
Model1.fit(xTrain,yTrain)
```

```python
XTr1=Model1.predict(xTrain)
TrainingAcc1=acs(yTrain,XTr1.round())
TrainingAcc1
```

```python
XTe1=Model1.predict(xTest)
TestingAc=acs(yTest,XTe1.round())
TestingAc
```
